```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Otsu vs other features

```python
df_otsu = pd.read_csv('otsu_masking.csv')
```

```python
# Plotting 
fig, ax = plt.subplots(5, 1, figsize=(4, 8))
c = 0
bands = ['g', 'r', 'i', 'z', 'Y']
cut_otsu = [0.003, 0.05, 0.14, 0.09, 0.09]
cut_area = [0.9, 0.7, 0.7, 0.8, 0.6]
for ax in ax:
    df_tmp = df_otsu.loc[df_otsu['band'] == bands[c]]
    # Number of selected
    Nsel = df_tmp.loc[(df_tmp['otsu'] >= cut_otsu[c]) & (df_tmp['area'] <= cut_area[c])]
    Nsel = len(Nsel.index)
    # Scatter plot
    kw = {
        's': 2,
        'color': 'navy',
        'label': '{0}, N={1}'.format(bands[c], Nsel),
    }
    ax.scatter(df_tmp['area'], df_tmp['otsu'], **kw)
    # Plot limits
    ax.axvline(cut_area[c], color='orange')
    ax.axhline(cut_otsu[c], color='orange')
    c += 1
    ax.legend(loc='upper right')
```
